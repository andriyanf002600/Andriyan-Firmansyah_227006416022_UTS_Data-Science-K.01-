```python
import pandas as pd

heart = pd.read_csv('data/heart/heart_disease.csv')
diabetes = pd.read_csv('data/diabetes/diabetes.csv')

heart.head(), diabetes.head()
```

evaluasi hasil data understanding

```python
# --- 1. Rename kolom diabetes agar konsisten ---
df_diabetes_fixed = df_diabetes.rename(columns={
    'age': 'Age',
    'chol': 'Cholesterol Level',
    'bp.1s': 'Blood Pressure',
    'gender': 'Gender'
})

# --- 2. Hitung BMI pada diabetes jika height & weight tersedia ---
df_diabetes_fixed['height'] = pd.to_numeric(df_diabetes_fixed['height'], errors='coerce')
df_diabetes_fixed['weight'] = pd.to_numeric(df_diabetes_fixed['weight'], errors='coerce')

df_diabetes_fixed['BMI'] = df_diabetes_fixed['weight'] / (df_diabetes_fixed['height']/100)**2

# --- 3. Drop kolom diabetes yang tidak relevan ---
df_diabetes_final = df_diabetes_fixed[['Age','Cholesterol Level','Blood Pressure','BMI','Gender']]

# --- 4. Ambil kolom dari heart disease agar sama ---
df_heart_final = df_heart[['Age','Cholesterol Level','Blood Pressure','BMI','Gender']]

# --- 5. Gabungkan kedua dataset ---
combined = pd.concat([df_heart_final, df_diabetes_final], ignore_index=True)

# --- 6. Buang baris yang kosong pada kolom utama ---
clean = combined[['Age','Cholesterol Level','Blood Pressure','BMI']].dropna()

# --- 7. Hitung korelasi ---
clean_corr = clean.corr()
clean_corr
```

# Cek Outlier

Metode Z-score

pilih hanya kolom numerik

```python
numeric_cols = combined.select_dtypes(include=['float64', 'int64'])
numeric_cols.head()
```

```python
from scipy import stats
import numpy as np

z = np.abs(stats.zscore(numeric_cols))
outliers = (z > 3).sum()

outliers
```

**Visualisasikan hasil analisis pada dataset anda dan sertakan alasan kenapa anda memilih chart tersebut!**

Histogram (Distribusi BMI)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ðŸ”¹ Cek dulu nama kolom yang tersedia
print(df_diabetes.columns)
print(df_heart.columns)
```

### Visualisasi: Distribusi Umur pada Dataset Diabetes

Saya menggunakan **histogram** untuk menampilkan distribusi umur pasien dalam dataset diabetes. Histogram ini menunjukkan jumlah pasien pada rentang umur tertentu, dengan garis KDE (Kernel Density Estimate) sebagai estimasi distribusi yang lebih halus.

ðŸ“Œ **Alasan pemilihan chart**:  
Histogram sangat cocok untuk data numerik seperti umur, karena membantu memahami **demografi usia** pasien. Dari grafik terlihat bahwa mayoritas pasien berada di sekitar usia 40 tahun, dengan jumlah yang menurun di atas usia 50 dan di bawah usia 30.

ðŸ“Œ **Insight dari visualisasi**:  
Distribusi ini bisa menjadi dasar analisis lebih lanjut, misalnya apakah usia tertentu memiliki risiko diabetes lebih tinggi, atau bagaimana strategi pencegahan bisa difokuskan pada kelompok usia dominan.


```python
# ðŸ”¹ 1. Histogram (numerik dari dataset diabetes â†’ kolom 'age')
sns.histplot(df_diabetes['age'], bins=20, kde=True)
plt.title("Distribusi Umur pada Dataset Diabetes")
plt.xlabel("Umur")
plt.ylabel("Jumlah Pasien")
plt.show()
```

### Visualisasi: Distribusi Tekanan Darah Tinggi pada Dataset Heart Disease

Saya menggunakan **bar chart** untuk menampilkan distribusi pasien berdasarkan status tekanan darah tinggi dalam dataset penyakit jantung. Grafik ini menunjukkan jumlah pasien yang memiliki tekanan darah tinggi (High Blood Pressure = Yes) dan yang tidak (No).

ðŸ“Œ **Alasan pemilihan chart**:  
Bar chart sangat cocok untuk data kategori seperti ini (Ya/Tidak), karena memudahkan kita melihat **proporsi dan perbandingan jumlah pasien** dalam setiap kategori.

ðŸ“Œ **Insight dari visualisasi**:  
Terlihat bahwa jumlah pasien dengan tekanan darah tinggi sedikit lebih banyak dibandingkan yang tidak. Hal ini menunjukkan bahwa tekanan darah tinggi merupakan faktor yang cukup dominan pada pasien penyakit jantung, dan bisa menjadi fokus utama dalam strategi pencegahan atau pengelolaan risiko.


```python
sns.countplot(x='High Blood Pressure', data=df_heart)
plt.title("Distribusi Tekanan Darah Tinggi pada Dataset Heart Disease")
plt.xlabel("High Blood Pressure (Ya/Tidak)")
plt.ylabel("Jumlah Pasien")
plt.show()
```

### Visualisasi: Heatmap Korelasi Variabel pada Dataset Diabetes

Saya memilih **heatmap korelasi** karena ingin melihat hubungan antar variabel numerik dalam dataset diabetes. Heatmap ini menunjukkan seberapa kuat korelasi antara dua variabel, dengan nilai berkisar dari -1 (korelasi negatif sempurna) hingga +1 (korelasi positif sempurna).

Beberapa korelasi yang menonjol:
- `weight` dan `waist` memiliki korelasi tinggi (0.87), menunjukkan bahwa berat badan sangat berkaitan dengan ukuran pinggang.
- `glyhb` dan `ratio` memiliki korelasi kuat (0.78), yang bisa menunjukkan hubungan antara kadar glikemik dan rasio kolesterol.
- `bp.1s` dan `bp.2s` (tekanan darah sistolik pertama dan kedua) juga sangat berkorelasi (0.83), menunjukkan konsistensi pengukuran tekanan darah.

ðŸ“Œ **Alasan pemilihan chart**:  
Heatmap dipilih karena efektif untuk menampilkan **hubungan antar indikator kesehatan** secara visual dan cepat. Ini membantu dalam memahami variabel mana yang saling berkaitan dan bisa menjadi fokus analisis lebih lanjut.



```python
numeric_df_diabetes = df_diabetes.select_dtypes(include=['number'])
corr = numeric_df_diabetes.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heatmap Korelasi Variabel pada Dataset Diabetes")
plt.show()
```

### Visualisasi: Proporsi Pasien Diabetes pada Dataset Heart Disease

Saya menggunakan **pie chart** untuk menampilkan proporsi pasien yang memiliki diabetes dan yang tidak dalam dataset penyakit jantung. Pie chart ini menunjukkan bahwa distribusi antara pasien diabetes dan non-diabetes hampir seimbang, yaitu sekitar 49.7% pasien diabetes dan 50.3% tidak.

ðŸ“Œ **Alasan pemilihan chart**:  
Pie chart sangat cocok untuk menampilkan **proporsi kategori** secara visual. Dalam konteks ini, kita bisa langsung melihat bahwa diabetes merupakan kondisi yang cukup umum di antara pasien penyakit jantung.

ðŸ“Œ **Insight dari visualisasi**:  
Distribusi yang hampir seimbang ini menunjukkan bahwa diabetes bisa menjadi faktor risiko penting dalam analisis penyakit jantung. Hal ini dapat menjadi dasar bagi rumah sakit atau lembaga kesehatan untuk merancang program pencegahan dan edukasi yang menyasar kedua kelompok secara proporsional.


```python
labels = ['Tidak Diabetes', 'Diabetes']
sizes = df_heart['Diabetes'].value_counts()
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Proporsi Pasien Diabetes pada Dataset Heart Disease")
plt.show()
```
